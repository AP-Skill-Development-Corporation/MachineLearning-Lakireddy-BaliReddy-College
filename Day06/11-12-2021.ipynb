**1. Palette**
- In seaborn palettes is ana interface to generate different color shades.
    - Qualitative
    - Sequential
    - Diverging

```python
### Qualitative Palette: Palette Contains Multiple colors
sns.palplot(sns.color_palette())
```

```python
help(sns.color_palette)
```

```python
sns.palplot(sns.color_palette('colorblind'))
```

```python
sns.palplot(sns.color_palette('pastel'))
```

```python
### Sequential: Single color palette with lite to bright colors
sns.palplot(sns.color_palette('Reds'))
```

```python
sns.palplot(sns.color_palette('Purples'))
```

```python
sns.palplot(sns.color_palette('Blues'))
```

```python
### Diverging: Palette contains 2 colors
sns.palplot(sns.color_palette('RdBu', n_colors = 100))
```

```python
sns.palplot(sns.color_palette('RdBu', n_colors = 15))
```

```python
sns.palplot(sns.color_palette('OrRd'))
```

```python
sns.palplot(sns.color_palette('PuOr'))
```

```python
sns.palplot(sns.dark_palette('Purple', reverse = True))
```

```python
sns.palplot(sns.color_palette('PuOrs'))
# Error and also gives you available colors
```

**Categorical Plot**
- sns.catplot()

```python
help(sns.catplot)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = df)
```

```python
sns.catplot(x = 'species', y = 'sepal_width', data = df, kind = 'box', 
            palette = 'gnuplot_r')
```

```python
df2 = sns.load_dataset('titanic')
df2.head()
```

```python
sns.catplot(x = 'sex', y = 'survived', data = df2, kind = 'bar')
```

```python
sns.catplot(x = 'sex', y = 'survived', data = df2, kind = 'bar', hue = 'class')
```

**Joint Plot**
- Combination of 2 plots
- sns.jointplot()

```python
help(sns.jointplot)
```

```python
sns.jointplot(x='petal_width', y='petal_length',data = df)
```

```python
sns.jointplot(x='petal_width', y='sepal_width',data = df, 
              kind = 'hex', color = 'orange')
```

**Pairplot**
- Pairwise relation between the columns
- sns.pairplot()

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'species', markers = ['o','+','v'], 
             palette = 'Spectral_r')
```

```python
sns.pairplot(df,hue='species',markers = ['o','+','v']) 
```
